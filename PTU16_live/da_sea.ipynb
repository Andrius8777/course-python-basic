```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
sns.get_dataset_names()
```

```python
tips = sns.load_dataset('tips')
tips.tail()
```

```python
tips.shape
```

```python
sns.displot(tips['total_bill'])
```

```python
sns.displot(data=tips['total_bill'], bins=4)
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips)
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex')
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='reg')
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='hist')
```

```python
sns.scatterplot(x='total_bill', y='tip', data=tips)
```

```python
sns.scatterplot(x='total_bill', y='tip', data=tips, hue='sex')
```

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips, hue='sex', diag_kind='hist')
```

```python
sns.scatterplot(tips, x='tip', y='size')
```

```python
sns.barplot(tips, x='sex', y='total_bill', estimator=sum)
```

```python
sns.barplot(tips, x='sex', y='total_bill', hue='day', estimator=sum, errorbar=('ci', False))
```

```python
sns.countplot(tips, x='sex', hue='smoker')
```

```python
sns.boxplot(tips, x='sex', y='total_bill')
```

```python
koerliacijos = tips.corr(numeric_only=True)
koerliacijos
```

```python
sns.heatmap(koerliacijos, annot=True, cmap='BuPu')
```
