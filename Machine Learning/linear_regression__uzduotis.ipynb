```python
fish.info()
```

```python
dummies = pd.get_dummies(fish['Species'])
```

```python
dummies.head(50)
```

```python
table_data = pd.concat([fish, dummies], axis = 1)
```

```python
table_data.head()
```

```python
table_data.drop(['Species'], axis=1, inplace=True)
```

```python
table_data.head()
```

```python
table_data.columns
```

```python
parametrai = table_data[['Length1', 'Length2', 'Length3', 'Height', 'Width', 'Bream',
       'Parkki', 'Perch', 'Pike', 'Roach', 'Smelt', 'Whitefish']]
```

```python
atsakymas = table_data['Weight']
```

```python
parametrai
```

```python
parametrai
```

```python
atsakymas
```

```python
from sklearn.model_selection import train_test_split
```

```python
parametrai_train, parametrai_test, atsakymas_train, atsakymas_test = train_test_split(parametrai, atsakymas, test_size=0.33, random_state=50)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
prediction = LinearRegression().fit(parametrai_train, atsakymas_train)
```

```python
prognozes = prediction.predict(parametrai_test)
```

```python
prognozes
```

```python
prediction.score(parametrai_test, atsakymas_test)
```

```python
spejimai = pd.Series(data=prognozes, name='Rezultatai')
```

```python
results = pd.concat([atsakymas_test.reset_index(), spejimai], axis=1)
```
